```python
import os
os.environ['TF_DETERMINISTIC_OPS'] = '1'

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Dense
from tensorflow.keras.initializers import RandomNormal
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Concatenate
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
from tensorflow.keras.initializers import HeNormal
from tensorflow.keras.layers import Conv2D, Flatten
from keras.initializers import HeUniform
import matplotlib.pyplot as plt
import json
import numpy as np
import math
import random
import array
import gc
from tensorflow.keras.backend import clear_session

def generate_midpoints(n):
     # Create a grid of size n+1 that spans the edges of the cells within [-1, 1]
    x = np.linspace(-1, 1, n + 1)
    y = np.linspace(-1, 1, n + 1)

    # Calculate the midpoints correctly within the specified range
    mid_x = (x[:-1] + x[1:]) / 2
    mid_y = ((y[:-1] + y[1:]) / 2)[::-1] # Reverse the y midpoints for top-to-bottom order

    # Generate the Cartesian grid midpoints
    midpoints = [(mx, my) for my in mid_y for mx in mid_x]

    return np.array(midpoints)

def generate_types(num_classes, num_samples):
    
    random_labels = np.random.randint(0, num_classes, size=num_samples)
    one_hot_encoded = to_categorical(random_labels, num_classes=num_classes)
    return one_hot_encoded

def generate_indicator(n):
    random_indicator = np.random.randint(0, 2, size=n)
    return random_indicator

def merge_arrays(arr1, arr2, arr3):
    merged = [(a, b, c) for a, b, c in zip(arr1, arr2, arr3)]
    return np.array(merged)



#funciones para procesar el output del player...

# no 
def swap_classes_based_on_value(one_hot, value_to_change):
    # Convierte el vector a un array de numpy si no lo es ya
    one_hot = np.array(one_hot)
    
    # Verifica cuál es el índice del valor 1
    idx = np.argmax(one_hot)
    
    # Si el índice es 0 (1,0,0) lo cambia a (0,1,0)
    if idx == 0 and value_to_change == 0:
        return np.array([0, 1, 0])
    # Si el índice es 1 (0,1,0) lo cambia a (1,0,0)
    elif idx == 1 and value_to_change == 0:
        return np.array([1, 0, 0])
    # Si el índice es 0 (1,0,0) lo cambia a (0,1,0) para 1
    elif idx == 0 and value_to_change == 1:
        return np.array([0, 1, 0])
    # Si el índice es 1 (0,1,0) lo cambia a (1,0,0) para 1
    elif idx == 1 and value_to_change == 1:
        return np.array([1, 0, 0])
    # Si el índice es 2 (0,0,1) lo deja igual
    else:
        #return np.array([1, 0, 0])
        return one_hot

def swap_classes_in_array_based_on_player(one_hot_array, control_array, value_to_change):
    return np.array([swap_classes_based_on_value(one_hot, value_to_change) if control == value_to_change else one_hot 
                     for one_hot, control in zip(one_hot_array, control_array)])


def concatenate_if_not_empty(array1, array2, axis=0):
    if array1.size > 0 and array2.size > 0:
        return np.concatenate((array1, array2), axis=axis)
    elif array1.size > 0:
        return array1
    elif array2.size > 0:
        return array2
    else:
        return np.array([])


def find_common_decimal_coordinates(array1, array2, decimals=2):
    if array1.size == 0 or array2.size == 0:
        return np.array([])

    # Multiplicar los elementos por 10^decimals y convertir a enteros
    factor = 10 ** decimals
    array1_scaled = (array1 * factor).astype(int)
    array2_scaled = (array2 * factor).astype(int)
    
    # Encontrar las entradas comunes
    common_entries = np.array([coord for coord in array1_scaled if any(np.array_equal(coord, x) for x in array2_scaled)])
    
    # Convertir de nuevo a la forma original y dividir por el factor
    common_entries = common_entries / factor
    
    return common_entries

    


def remove_common_entries(array1, array2, *other_arrays):
    if array1.size == 0 or array2.size == 0:
        return array1, other_arrays

    # Multiplicar los elementos por 10^decimals y convertir a enteros
    factor = 100  # para comparar hasta 2 decimales
    array1_scaled = (array1 * factor).astype(int)
    array2_scaled = (array2 * factor).astype(int)

    # Convertir array2_scaled a un conjunto de tuplas para búsqueda rápida
    array2_set = set(map(tuple, array2_scaled))

    # Encontrar los índices de los valores en array1 que están en array2
    indices_to_remove = [i for i, coord in enumerate(array1_scaled) if tuple(coord) in array2_set]

    #half = len(indices_to_remove) // 2
    #indices_to_remove = indices_to_remove[half:]

    # Convertir indices_to_remove a un array NumPy
    indices_to_remove = np.array(indices_to_remove)

    if indices_to_remove.size == 0:
        return [array1] + list(other_arrays)

    # Eliminar los valores en esos índices de array1
    array1_filtered = np.delete(array1, indices_to_remove, axis=0)

    # Eliminar los valores en esos índices de los otros arrays
    other_arrays_filtered = [np.delete(arr, indices_to_remove, axis=0) for arr in other_arrays]

    # Imprimir el número de valores eliminados
    print(f"Se eliminaron {len(indices_to_remove)} valores de array1.")
    
    return [array1_filtered] + other_arrays_filtered



def modify_indicator(matrix):
    new_array= np.zeros(matrix.shape[0], dtype=int)
    for i, onehot_vector in enumerate (matrix):
        if np.argmax(onehot_vector) == 0:
            new_array[i] = 0
        elif np.argmax(onehot_vector) == 1:
            new_array[i] = 1
    return new_array

def save_output(folder_name, i):
  comando_cpnm = f"cp ./training-envs-executables/linux/MechanicMiner04/MechanicMiner04_Data/logs_player.json {folder_name}/nm{i}.json" #NEW MECHANIC!!!!
  comando_cpom = f"cp ./training-envs-executables/linux/MechanicMiner02/MechanicMiner02_Data/logs_player.json {folder_name}/om{i}.json"  # OLD MECHANIC
  os.system(comando_cpnm)
  os.system(comando_cpom)

def process_enemy_output(folder_name, i):
    
  #comando_cpnm = f"cp ./training-envs-executables/linux/MechanicMiner01/MechanicMiner01_Data/logs_player.json {folder_name}/nm{i}.json"
  #comando_cpom = f"cp ./training-envs-executables/linux/MechanicMiner10/MechanicMiner10_Data/logs_player.json {folder_name}/om{i}.json"
  #os.system(comando_cpnm)
  #os.system(comando_cpom)

  nm_file_path = f"{folder_name}/nm{i}.json"
  om_file_path = f"{folder_name}/om{i}.json"
  enemy_matrix_path = f"{folder_name}/original.csv"  

    
  with open(nm_file_path, 'r') as file: 
      datanm = json.load(file)
    
  with open(om_file_path, 'r') as file:
      dataom = json.load(file)
    
  with open(enemy_matrix_path, 'r') as file:
      enemy = file.read()
  # necesito cambiar esto para que lo lea de los logs   
  #with open('./training-envs-executables/linux/MechanicMiner01/MechanicMiner01_Data/logs_player.json', 'r') as file:
      #datanm = json.load(file)
  #with open('./training-envs-executables/linux/MechanicMiner10/MechanicMiner10_Data/logs_player.json', 'r') as file:
      #dataom = json.load(file)
  #with open('./training-envs-executables/linux/MechanicMiner01/MechanicMiner01_Data/thresholded_matrix.csv', 'r') as file:
    #enemy = file.read()
    

  # Extract input and output from the JSON data
  outputsnm = datanm['times_won']
  outputsom = dataom['times_won']

  mee = datanm['total_times_won']
  mee2 = dataom['total_times_won']  
    
  print("TIMES WON NM: ",mee)
  print("TIMES WON OM: ",mee2)

  #new mechanics labels...
  #temp_playerDir_om['local_position_player']['direction_player']
  playerDir_nm = datanm['local_position_player'] #position of the player given the enemy.
  playerposition_nm = datanm['enemyPosition'] #position that was hit in the matrix.
  onehot_nm = datanm['oneHotEncoding'] # part of the enemy that was hit.
  agent_nm =  np.ones(len(playerposition_nm)) 

  #old mechanics labels
  playerDir_om = dataom['local_position_player']
  playerposition_om = dataom['enemyPosition']   ### this is enemy positon not player p
  onehot_om = dataom['oneHotEncoding']  
  agent_om =  np.zeros(len(playerposition_om))


  #to list..
  player_position_nm = [list(d.values()) for d in playerposition_nm]
  player_direction_nm =  [list(d.values()) for d in playerDir_nm]
  result_onehot_nm= [[int(num) for num in encoding.split(",")] for encoding in onehot_nm] 

  player_position_om = [list(d.values()) for d in playerposition_om]
  player_direction_om =  [list(d.values()) for d in playerDir_om]
  result_onehot_om= [[int(num) for num in encoding.split(",")] for encoding in onehot_om] 
  

    
  #swap here  
  # 3 arrays 1 position of the player 2 one hot encode on that position of collision 3 the one-hot output that should be for place (Y) label? 
  #player_position_nm = [list(d.values()) for d in temp_playerDir_nm]
  #player_position_nm = [list(d.values()) for d in temp_playerposition_nm]
  #player_direction_nm =  [list(d.values()) for d in temp_playerDir_nm]
  #result_onehot_nm= [[int(num) for num in encoding.split(",")] for encoding in temp_onehot_nm] #get_values_from_coordinates(player_position_nm,one_hot_vectors,4)
  #print(result_onehot_nm)
  result_output_nm = swap_classes_in_array_based_on_player(result_onehot_nm,outputsnm,0) # descomentado 2
  print("__________________________________________")
  #print(result_output_nm)
  #result_output_nm = result_onehot_nm # des comentado 1
  #print(result_onehot_nm[0],result_output_nm[0])
    
  #player_position_om = [list(d.values()) for d in temp_localplayerpositionom]
  #player_position_om = [list(d.values()) for d in temp_localplayerpositionom]
  #player_direction_om =  [list(d.values()) for d in temp_playerDir_om]
  #result_onehot_om=    [[int(num) for num in encoding.split(",")] for encoding in temp_onehot_om]#get_values_from_coordinates(player_position_om,one_hot_vectors,4)
  result_output_om = swap_classes_in_array_based_on_player(result_onehot_om,outputsom,1)

  print(result_onehot_om[0],result_output_om[0])  


  print("__________________________________________")
  #print(result_output_om)   
  # Convert Python lists to NumPy arrays
  player_position_nm = np.array(player_position_nm)
  player_position_om = np.array(player_position_om)
  player_direction_nm = np.array(player_direction_nm)
  player_direction_om = np.array(player_direction_om)
  result_onehot_nm = np.array(result_onehot_nm)
  result_onehot_om = np.array(result_onehot_om)
  outputsnm = np.array(outputsnm)
  outputsom = np.array(outputsom)
  result_output_nm = np.array(result_output_nm)
  result_output_om = np.array(result_output_om)

  #evaluate the things before concatenate the valores...
  print("total de todos los valores",len(player_position_nm))
  results = remove_common_entries(player_position_nm,player_position_om,result_onehot_nm,outputsnm,result_output_nm,agent_nm)

  player_position_nm =  results[0]
  result_onehot_nm   =  results[1]
  outputsnm          =  results[2]
  result_output_nm   =  results[3]
  agent_nm           =  results[4]

  print("total de todos los valores despues de quitar",len(player_position_nm))
  reduceom = len(player_position_nm)

  player_position_om =  player_position_om[reduceom:]
  result_onehot_om = result_onehot_om[reduceom:]
  outputsom = outputsom[reduceom:]
  result_output_om = result_output_om[reduceom:]
  agent_om = agent_om[reduceom:]
  
  print("total de todos los valores del otro valor om",len(player_position_om))

  player_position = concatenate_if_not_empty(player_position_nm, player_position_om)
  #player_direction = concatenate_if_not_empty(player_direction_nm, player_direction_om)
  one_hot_from_position = concatenate_if_not_empty(result_onehot_nm, result_onehot_om)
  result_of_collision = concatenate_if_not_empty(outputsnm, outputsom)
  new_labels_onehot = concatenate_if_not_empty(result_output_nm, result_output_om)
  agent = concatenate_if_not_empty(agent_nm, agent_om)
  print("suma de todos")
  print(len(player_position))
  #print(len(player_direction))
  print(len(one_hot_from_position))
  print(len(new_labels_onehot))
  print(len(agent))
  print("total of collisions:",len(result_of_collision))  
  #player_position =  np.concatenate((player_position_nm, player_position_om), axis=0) # player position new and old mechanic
  #one_hot_from_position =  np.concatenate((result_onehot_nm, result_onehot_om), axis=0) #all the original one-hot encode positions of the enemy
  #result_of_collision =  np.concatenate((temp_outputsnm, temp_outputsom), axis=0) #result of the collision on the simulation 0 loss 1 won 
  #new_labels_onehot =  np.concatenate((result_output_nm, result_output_om), axis=0) # new one-hot encodes that will be used for training...

  #shuffle data...   
  arrays = [player_position, one_hot_from_position, result_of_collision, new_labels_onehot, agent] 
  n_elements = len(player_position)
  permutation = np.random.permutation(n_elements)

  for i, arr in enumerate(arrays):
      if arr.ndim == 1:
          arrays[i] = arr[permutation]
      elif arr.ndim == 2:
          arrays[i] = arr[permutation, :]
      #player_position, one_hot_from_position, result_of_collision, new_labels_onehot, agent
  return arrays[0],arrays[1],arrays[2],arrays[3],arrays[4]
```

```python
def setting_seed(index):
    os.environ['TF_DETERMINISTIC_OPS'] = '1'
    os.environ['PYTHONHASHSEED'] = '0'
    # Set the seed for reproducibility
    seed =values[index]
    np.random.seed(seed)
    tf.random.set_seed(seed)
    random.seed(seed)
    he_uniform_initializer = HeUniform(seed=seed)

    return he_uniform_initializer

#

```

```python
he_uniform_initializer = setting_seed(0)

# Definir las entradas
one_hot_input = Input(shape=(3,), name='one_hot_input')
coordinates_input = Input(shape=(2,), name='coordinates_input')  # Supongamos que tenemos coordenadas (x, y)
collision_input = Input(shape=(1,), name='collision_input')  # collision input
#agent_input = Input(shape=(1,), name='agent_input')  # agent input

# Concatenar las entradas por columnas (axis=-1)
concatenated = Concatenate(axis=-1)([one_hot_input, coordinates_input, collision_input])#, collision_input,agent_input])

#inicializar ocn seed 

# Definir las capas densas
dense1 = Dense(12, activation='relu', kernel_initializer=he_uniform_initializer)(concatenated)
dense2 = Dense(6, activation='relu', kernel_initializer=he_uniform_initializer)(dense1)
output = Dense(3, activation='softmax', kernel_initializer=he_uniform_initializer)(dense2)

# Definir el modelo
model = Model(inputs=[one_hot_input, coordinates_input, collision_input], outputs=output) #model = Model(inputs=[one_hot_input, coordinates_input, collision_input,agent_input], outputs=output)

# Compilar el modelo
model.compile(optimizer='adam',
              loss= 'categorical_crossentropy',  # Mean Squared Error para problemas de regresión categorical_crossentropy wasserstein_loss  mse
              metrics=['accuracy'])




#model.summary()
#model.fit([midpoints, one_hot_data, indicator], labels, epochs=10, batch_size=32)
#print(types)
#print(midpoints)
#print(indicator)
predictions = model.predict([types, midpoints, indicator])
#print(predictions)

# Get index of the maximum probability class
predicted_class = np.argmax(predictions, axis=-1)
f_output = np.zeros_like(predictions)
f_output[np.arange(len(predictions)), predicted_class] = 1
print("resultado final???")
print(f_output)

#types = f_output
#indicator = modify_indicator(types)

np.savetxt('thresholded_matrix.csv', f_output.reshape(1, -1), delimiter=',', fmt='%d')
```

```python
def remove_common_entries(array1, array2, *other_arrays):
    if array1.size == 0 or array2.size == 0:
        return array1, other_arrays

    # Multiplicar los elementos por 10^decimals y convertir a enteros
    factor = 100  # para comparar hasta 2 decimales
    array1_scaled = (array1 * factor).astype(int)
    array2_scaled = (array2 * factor).astype(int)

    # Encontrar los índices de los valores en array1 que están en array2
    indices_to_remove = [i for i, coord in enumerate(array1_scaled) if any(np.array_equal(coord, x) for x in array2_scaled)]
    
    # Eliminar los valores en esos índices de array1
    array1_filtered = np.delete(array1, indices_to_remove, axis=0)

    # Eliminar los valores en esos índices de los otros arrays
    other_arrays_filtered = [np.delete(arr, indices_to_remove, axis=0) for arr in other_arrays]

    # Imprimir el número de valores eliminados
    print(f"Se eliminaron {len(indices_to_remove)} valores de array1.")
    
    return array1_filtered, other_arrays_filtered
```
